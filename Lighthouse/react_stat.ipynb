```python
import pandas as pd

# Load CSV data into DataFrame
file_path = '/Users/nirmalkapilarathne/Desktop/lighthouse_dekstop/React/react_all_performance_metrics.csv'
df = pd.read_csv(file_path)

# Remove non-numeric characters and convert to numeric
df = df.replace({'\xa0': '', 's': '', 'ms': ''}, regex=True).apply(
    pd.to_numeric, errors='coerce')

# Calculate descriptive statistics
statistics = {
    'Mean': df.mean(),
    'Median': df.median(),
    'Maximum': df.max(),
    'Minimum': df.min(),
    'Standard Deviation': df.std()
}

# Convert statistics to DataFrame
statistics_df = pd.DataFrame(statistics)

# Save statistics to Excel file
output_file_path = '/Users/nirmalkapilarathne/Desktop/lighthouse_dekstop/React/react_statistics.xlsx'
statistics_df.to_excel(output_file_path)

# Display the calculated statistics
statistics_df
```
